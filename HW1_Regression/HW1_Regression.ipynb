```python
from sklearn.model_selection import train_test_split

features =['bathrooms','sqft_living', 'grade', 'sqft_above', 'sqft_living15']    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)

print("number of training samples:",x_train.shape[0])
print("number of test samples:", x_test.shape[0])
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)

lr.score(x_test, y_test)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(Y, lr.predict(X))
print('R^2: ', r2)
```

```python
features =['bedrooms', 'bathrooms', 'sqft_living', 'view', 'grade', 
          'sqft_above', 'sqft_basement', 'lat', 'sqft_living15'] 

X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)

lr.score(x_test, y_test)
```

```python
r2 = r2_score(Y, lr.predict(X))
print('R^2: ', r2)
```

```python
plt.scatter(y_test, lr.predict(x_test))
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2)

x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

X_poly = poly.fit_transform(X)
```

```python
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(x_train_poly, y_train)
```

```python
lin_reg_poly.score(x_test_poly, y_test)
```

```python
r2_score(Y, lin_reg_poly.predict(X_poly))
```

```python
plt.scatter(y_test, lin_reg_poly.predict(x_test_poly))
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
y_pred = ridge.predict(x_test)

print(ridge.score(x_test, y_test))

r2 = r2_score(y_test, y_pred)
print('R^2:', r2)
```

```python
plt.scatter(y_test, ridge.predict(x_test))
plt.xlabel('True Price')
plt.ylabel('Predicted Price')
```
